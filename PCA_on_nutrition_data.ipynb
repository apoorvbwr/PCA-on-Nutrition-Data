# Explore the data distributions

```python
ax = data.hist(bins = 50, xlabelsize = -1, ylabelsize = -1, figsize = (11,11))
```

```python
from scipy.stats import boxcox
# add 1 because data must be positive (we have many zeros)
data = data + 1
data_tf = pd.DataFrame(index=data.index)
for i in data.columns.values:
    data_tf["%s_TF" % i] = boxcox(data.loc[:, i])[0]
```

```python
ax = data_tf.hist(bins=50, xlabelsize=-1, ylabelsize=-1, figsize=(11,11))
```

To account for different scales of measurement, we'll standardize to mean = 0 and variance = 1

```python
data_tf = StandardScaler().fit_transform(data_tf)
print('mean: ',np.round(data_tf.mean()))
```

```python
print('standard dev: ', np.round(data_tf.std()))
```

# Implement PCA

```python
fit = PCA()
```

```python
pca = fit.fit_transform(data_tf)
```

```python
plt.plot(fit.explained_variance_ratio_)
```

```python
print(fit.explained_variance_ratio_)
```

```python
pca = pd.DataFrame(pca[:, :5], index=data.index)
pca = pca.join(data_desc)
pca.drop(['CommonName','MfgName','ScientificName'], axis=1, inplace=True)
pca.rename(columns={0:'c1',1:'c2',2:'c3',3:'c4',4:'c5'}, inplace=True)
```
